```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
```

```python
target = new_client.target
activity = new_client.activity

# Example: EGFR
target_query = target.search("HDAC1")
targets = pd.DataFrame.from_dict(target_query)
targets

```

```python
HDAC = targets.target_chembl_id[2]
print("target ChembL ID", HDAC)
results = activity.filter(target_chembl_id=HDAC, standard_type="IC50")

data = pd.DataFrame(list(results))
```
